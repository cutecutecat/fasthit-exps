```python
data_dir = "../data/gb1/"
assert isdir(data_dir)

input_dir = "../tmp/"
assert isdir(input_dir)
```

```python
measured_data = pd.read_csv(
    os.path.join(data_dir, "elife-16965-supp1-v4.csv")
)
data = measured_data.loc[:, ["Variants", "Fitness"]]
score = data.loc[:, "Fitness"]# WT score is set to 1.

data.loc[:, "Fitness"] = (score - score.min()) / (score.max() - score.min())
data
```

```python
testset = data.sample(n=384, random_state=200)
testset.to_csv(os.path.join(input_dir, "testset.csv"), index=False)
testset
```

```python
trainset = data.drop(testset.index)
trainset.to_csv(os.path.join(input_dir, "trainset.csv"), index=False)
trainset
```
